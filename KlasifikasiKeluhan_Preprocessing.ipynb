# _Pre-processing_

```python
import pandas as pd
import re
import string

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
data = pd.read_csv('data/keluhan.csv', encoding='ISO-8859-1')
data = data.loc[0:len(data) - 2, 'Tweet':'Other']
data['Keluhan'] = data['Keluhan'].map({'Ya':1}).fillna(0)
data['Respon'] = data['Respon'].map({'Ya':1}).fillna(0)
data['Other'] = data['Other'].map({'Ya':1}).fillna(0)
data.head(10)
```

```python
def url_remove(tweet):
    pattern = re.compile("http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+")
    return re.sub(pattern, '', tweet)

def mention_remove(tweet):
    pattern = re.compile("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)")
    return re.sub(pattern, '', tweet)

def punctuation_remove(tweet):
    chars = set(string.punctuation)
    return ''.join(w for w in tweet if w not in chars)

def stopwords_remove(tweet):
    remover = StopWordRemoverFactory().create_stop_word_remover()
    return remover.remove(tweet)

def stem(tweet):
    stemmer = StemmerFactory().create_stemmer()
    return stemmer.stem(tweet)
```

```python
preprocessed = []
X = data['Tweet']

for i in range(len(X)):
    preprocessed.append(stopwords_remove(punctuation_remove(mention_remove(url_remove(X[i])))))
    
X = preprocessed
```

```python
preprocessed = []
print(X[0:10])

for i in range(len(X)):
    preprocessed.append(stem(X[i]))

X = preprocessed
print(preprocessed[0:10])
```
